```python
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df.corr()
```

```python
corr_mat=df.corr().abs()
plt.figure(figsize=[20,15])
sns.heatmap(corr_mat,annot=True, annot_kws={'size':10})
plt.title("correlation Matrix")
plt.show()
```

```python
df.corr()["W"].sort_values(ascending=False).abs()
```

```python
df=df.drop(columns=["ER"])
df=df.drop(columns=["ERA"])
df.head()
```

```python
df.shape
```

```python
df=df.drop(columns=["E"])
df=df.drop(columns=["CG"])
df=df.drop(columns=["H"])
df=df.drop(columns=["AB"])
df.head()
```

```python
plt.figure(figsize=(20,15))
data=['R', '2B','3B','HR','BB','SO','SB','RA','SHO','SV']

for i in data:
    sns.displot(df[i])
```

```python
x=df.drop(columns=['W'])
y=df['W']
```

```python
x
```

```python
y
```

```python
plt.figure(figsize=(15,10))
plotnumber=1

for column in x:
    if plotnumber <=12:
        ax=plt.subplot(3,4,plotnumber)
        plt.scatter(x[column],y)
        plt.xlabel(column,fontsize=10)
        plt.ylabel('w',fontsize=10)
        
    plotnumber+=1
plt.tight_layout()
    
```

```python
sc=StandardScaler()
x_scaled=sc.fit_transform(x)
```

```python
x_scaled
```

```python
x_train,x_test,y_train,y_test=train_test_split(x_scaled,y,test_size=0.30,random_state=6564)
y_train.head()
```

```python
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
lr.score(x_train,y_train)
```

```python
lr.score(x_test,y_test)
```

```python
y_pred=lr.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
mean_absolute_error(y_test,y_pred)
```

```python
mean_squared_error(y_test,y_pred)
```

```python
np.sqrt(mean_squared_error(y_test,y_pred))
```
